```python
people=[]
following_n=[]
followed_n=[]
following_list=[]
followed_list=[]
commit_n=[]
locations=[]
type_list=[]
bio_list=[]
```

```python
for i in range(len(filename)):
    file_path = os.path.join('data800', filename[i])
    data = json.load(open(file_path, encoding='utf-8'))
    print(file_path)
    personid = data['id']
    following=data['following_list']
    followed=data['follower_list']
    following_num=data['following']
    followed_num=data['followers']
    commits=data['commits']
    location=data['location']
    type1=data['type']
    bio=data['bio']
    
    people.append(personid)
    following_n.append(following_num)
    followed_n.append(followed_num)
    following_list.append(following)
    followed_list.append(followed)
    commit_n.append(commits)
    locations.append(location)
    type_list.append(type1)
    bio_list.append(bio)
```

```python
import pandas as pd

df=pd.DataFrame({'id':people,'following_n':following_n, 'followed_n':followed_n, 'following_list':following_list,'followed_list':followed_list,'commit_n':commit_n,'locations':locations,'type_list':type_list,'bio':bio_list}).fillna(0)
```

```python
all_id=[]
for i in range(len(df['id'])):
    all_id=all_id+df['following_list'][i]
    all_id=all_id+df['followed_list'][i]
```

```python
id0=[]
for i in range(len(all_id)):
    if all_id[i] in id0:
        continue
    else:
        id0.append(all_id[i])
```

```python
len(id0)
```

```python
link_list=[]
for i in range(len(df['id'])):
    tmp=df['id'][i]
    fol=df['following_list'][i]
    fol1=df['followed_list'][i]
    if fol!=[]:
        for i in range(len(fol)):
            link_list.append([str(tmp),str(fol[i])])
    if fol1!=[]:
        for i in range(len(fol1)):
            link_list.append([str(fol1[i]),str(tmp)])
```

```python
len(link_list)
```

```python
link_list[0]
```

```python
link=[]
for i in range(len(link_list)):
    tmpp=link_list[i]
    tmp={"source": tmpp[0], "target": tmpp[1], "value": 1}
    link.append(tmp)
```

```python
link
```

```python
len(link_list)
```

```python
ids=[]
for i in range(len(link_list)):
    l=link_list[i]
    ids.append(l[0])
    ids.append(l[1])
```

```python
idss=[]
for i in range(len(ids)):
    if ids[i] in idss:
        continue
    else:
        idss.append(ids[i])
```

```python
len(idss)
```

```python
idss
```

```python
node=[]
for i in range(len(idss)):
    idd=idss[i]
    tmp={"id":str(idd),"group":1}
    node.append(tmp)
```
